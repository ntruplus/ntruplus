```python
q = 3457
Xs = ND.CenteredBinomial(1)
Xe = ND.CenteredBinomial(1)
```

```python
NTRUPLUS576_LWE  = schemes.LWEParameters(n=576,  q=q, Xs=Xs, Xe=Xe, m=576,  tag='NTRU+576')
NTRUPLUS768_LWE  = schemes.LWEParameters(n=768,  q=q, Xs=Xs, Xe=Xe, m=768,  tag='NTRU+768')
NTRUPLUS864_LWE  = schemes.LWEParameters(n=864,  q=q, Xs=Xs, Xe=Xe, m=864,  tag='NTRU+864')
NTRUPLUS1152_LWE = schemes.LWEParameters(n=1152, q=q, Xs=Xs, Xe=Xe, m=1152, tag='NTRU+1152')
```

```python
NTRUPLUS576_NTRU  = schemes.NTRUParameters(n=576,  q=q, Xs=Xs, Xe=Xe, m=576,  tag='NTRU+576')
NTRUPLUS768_NTRU  = schemes.NTRUParameters(n=768,  q=q, Xs=Xs, Xe=Xe, m=768,  tag='NTRU+768')
NTRUPLUS864_NTRU  = schemes.NTRUParameters(n=864,  q=q, Xs=Xs, Xe=Xe, m=864,  tag='NTRU+864')
NTRUPLUS1152_NTRU = schemes.NTRUParameters(n=1152, q=q, Xs=Xs, Xe=Xe, m=1152, tag='NTRU+1152')
```

```python
print("=====================Classical_LWE=====================")
r = LWE.estimate(NTRUPLUS768_LWE, red_cost_model = RC.ADPS16)
print("=====================Classical_NTRU=====================")
r = NTRU.estimate(NTRUPLUS768_NTRU, red_cost_model = RC.ADPS16)
print("======================Quantum_LWE======================")
r = LWE.estimate(NTRUPLUS768_LWE, red_cost_model = RC.ChaLoy21)
print("======================Quantum_NTRU======================")
r = NTRU.estimate(NTRUPLUS768_NTRU, red_cost_model = RC.ChaLoy21)
```
